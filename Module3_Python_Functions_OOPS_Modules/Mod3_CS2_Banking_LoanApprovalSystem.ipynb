```python
import csv

# Read the CSV file and build a set of unique professions
with open('bank-data.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip header
    professions = set()
    ages = []
    for row in reader:
        profession = row[1]
        age = int(row[0])
        professions.add(profession.lower())  # Make profession check case insensitive
        ages.append(age)

# Compute max and min age for loan eligibility
max_age = max(ages)
min_age = min(ages)

# Store max and min age in a dictionary
age_range = {'max_age': max_age, 'min_age': min_age}

# Take input in a while loop until user types "END" for profession
while True:
    print("Max Age for Loan Eligibility: ", age_range['max_age'])
    print("Min Age for Loan Eligibility: ", age_range['min_age'])
    try:
        age=int(input("Enter age: "))
    except:
        print("Invalid input")
    else:
        profession = input("Enter a profession (or 'END' to exit): ").lower()
        if profession == 'end':
            break
        elif profession in professions and age >=min_age and age <=max_age:
            print("Client is eligible for marketing campaign.")

        else:
            print("Client is not eligible for marketing campaign.")
```
